```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
microbes = pd.read_csv('./data/microbes.tsv', sep='\t', index_col=0)
metabolites = pd.read_csv('./data/metabolites.tsv', sep='\t', index_col=0)
```

```python
print(f"min: {microbes.values.min()}, median: {np.median(microbes.values.ravel())}, max: {microbes.values.max()}, mean: {microbes.values.mean()}")
# print(np.sort(microbes.values.ravel())[-60:])
print(f"num_zero = {len(microbes.values.ravel()[microbes.values.ravel() == 0])} / {len(microbes.values.ravel())}")
```

```python
print(f"min: {metabolites.values.min()}, median: {np.median(metabolites.values.ravel())}, max: {metabolites.values.max()}, mean: {metabolites.values.mean()}")
# print(np.sort(metabolites.values.ravel())[-60:])
print(f"num_zero = {len(metabolites.values.ravel()[metabolites.values.ravel() == 0])} / {len(metabolites.values.ravel())}")
```
