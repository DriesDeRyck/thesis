```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from src.util import clr
```

```python
microbes = pd.read_csv('./data/microbes.tsv', sep='\t', index_col=0)
metabolites = pd.read_csv('./data/metabolites.tsv', sep='\t', index_col=0)
```

```python
microbes_scaled = microbes/microbes.values.mean()
metabolites_scaled = metabolites/metabolites.values.mean()
```

```python
print(np.log(np.nextafter(0, 1)))
```

```python
# replace zero's with smallest non-zero value in data
print(microbes.values[microbes.values != 0].min())
print(metabolites.values[metabolites.values != 0].min())
print(microbes_scaled.values[microbes_scaled.values != 0].min())
print(metabolites_scaled.values[metabolites_scaled.values != 0].min())

microbes_min = microbes.replace(0, microbes.values[microbes.values != 0].min())
metabolites_min = metabolites.replace(0, metabolites.values[metabolites.values != 0].min())
microbes_sc_min = microbes_scaled.replace(0, microbes_scaled.values[microbes_scaled.values != 0].min())
metabolites_sc_min = metabolites_scaled.replace(0, metabolites_scaled.values[metabolites_scaled.values != 0].min())

# replace all zero's with epsilon value
microbes.replace(0, np.nextafter(0, 1), inplace=True)
metabolites.replace(0, np.nextafter(0, 1), inplace=True)
microbes_scaled.replace(0, np.nextafter(0, 1), inplace=True)
metabolites_scaled.replace(0, np.nextafter(0, 1), inplace=True)
```

```python
concat = pd.concat([microbes, metabolites])
concat_scaled = pd.concat([microbes_scaled, metabolites_scaled])
concat_min = pd.concat([microbes_min, metabolites_min])
concat_sc_min = pd.concat([microbes_sc_min, metabolites_sc_min])
```

```python
pd.concat([np.log(concat).mean(axis=0), np.log(concat_scaled).mean(axis=0)], axis=1)
```

```python
clr_df = np.log(concat) - np.log(concat).mean(axis=0)
print(clr_df.values.mean())
clr_scaled_df = np.log(concat_scaled) - np.log(concat_scaled).mean(axis=0)
print(clr_scaled_df.values.mean())
clr_min_df = np.log(concat_min) - np.log(concat_min).mean(axis=0)
print(clr_min_df.values.mean())
clr_sc_min_df = np.log(concat_sc_min) - np.log(concat_sc_min).mean(axis=0)
print(clr_sc_min_df.values.mean())
```

```python
clr_microbes = clr_df.loc[microbes.index]
clr_metabolites = clr_df.loc[metabolites.index]
clr_scaled_microbes = clr_scaled_df.loc[microbes.index]
clr_scaled_metabolites = clr_scaled_df.loc[metabolites.index]
clr_min_microbes = clr_min_df.loc[microbes.index]
clr_min_metabolites = clr_min_df.loc[metabolites.index]
clr_sc_min_microbes = clr_sc_min_df.loc[microbes_sc_min.index]
clr_sc_min_metabolites = clr_sc_min_df.loc[metabolites_sc_min.index]
```

```python
clr_sc_min_metabolites.head()
```

```python
clr_microbes.to_csv('./data/clr_microbes.tsv', sep='\t', index=True)
clr_metabolites.to_csv('./data/clr_metabolites.tsv', sep='\t', index=True)
clr_scaled_microbes.to_csv('./data/clr_microbes_scaled.tsv', sep='\t', index=True)
clr_scaled_metabolites.to_csv('./data/clr_metabolites_scaled.tsv', sep='\t', index=True)
clr_min_microbes.to_csv('./data/clr_microbes_min.tsv', sep='\t', index=True)
clr_min_metabolites.to_csv('./data/clr_metabolites_min.tsv', sep='\t', index=True)
clr_sc_min_microbes.to_csv('./data/clr_sc_microbes_min.tsv', sep='\t', index=True)
clr_sc_min_metabolites.to_csv('./data/clr_sc_metabolites_min.tsv', sep='\t', index=True)
```
