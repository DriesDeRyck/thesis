```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from random import sample
import random
random.seed(4131)
```

```python
microbes = pd.read_csv('./data/microbes.tsv', sep='\t', index_col=0)
metabolites = pd.read_csv('./data/metabolites.tsv', sep='\t', index_col=0)
microbes.head()
```

```python
sample(list(microbes.columns), 10)
```

```python
subset20 = microbes[sample(list(microbes.columns), 20)]
subset100 = microbes[sample(list(microbes.columns), 100)]
subset500 = microbes[sample(list(microbes.columns), 500)]
```

```python
subset20.head()
```

```python
for i in [20, 100, 500]:
    columns_subset = sample(list(microbes.columns), i)
    microbes_subset = microbes[columns_subset]
    metabolites_subset = metabolites[columns_subset]
    microbes_subset.to_csv(f'./data/microbes.subset{str(i)}.tsv', sep='\t', index=True)
    metabolites_subset.to_csv(f'./data/metabolites.subset{str(i)}.tsv', sep='\t', index=True)
```
