```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
microbes_df = pd.read_csv("microbes.from_biom.tsv", sep='\t', index_col=0).drop(columns="9hr_late")
metabolites_df = pd.read_csv("metabolites.from_biom.tsv", sep='\t', index_col=0)
```

```python
metabolites_df.describe()
```

```python
metabolites_sum = metabolites_df.sum(axis=0, numeric_only=True)
microbes_sum = microbes_df.sum(axis=0, numeric_only=True)
metabolites_sum.plot()
```

```python
microbes_mean = np.nanmean(microbes_df)
microbes_std = np.nanstd(microbes_df)

metabolites_mean = np.nanmean(metabolites_df)
metabolites_std = np.nanstd(metabolites_df)
print(metabolites_mean)
```

```python
microbes_df.plot(kind='hist', logy=True, bins=20, subplots=True, layout=(5,4), figsize=(12, 9))
# plt.show()
plt.savefig("microbes_original.png")
```
