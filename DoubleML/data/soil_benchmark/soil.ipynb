```python
def plot_coefficient_matrix(coefficient_matrix):
    # Create figure
    figure = plt.figure(figsize=(10, 10))
    axes = figure.add_subplot()

    all_values = coefficient_matrix.values.ravel()
    vmin, vmax = -max(abs(all_values.min()), abs(all_values.max())), max(abs(all_values.min()), abs(all_values.max()))

    # Display the correlation matrix
    caxes = axes.matshow(coefficient_matrix, cmap='bwr', vmin=vmin, vmax=vmax)

    # Add colorbar
    figure.colorbar(caxes, orientation='horizontal')

    # Define labels for every 100 columns
    xlabels = coefficient_matrix.index
    ylabels = coefficient_matrix.columns
    # xticks = [i*100 + 50 for i in range(len(xlabels))]  # Approximate center of each partition
    # yticks = [i*100 + 50 for i in range(len(ylabels))]  # Approximate center of each partition
    #
    # # Apply labels to x and y axes
    # axes.set_xticks(xticks)
    # axes.set_yticks(yticks)
    # axes.set_xticklabels(xlabels, fontsize=12)
    # axes.set_yticklabels(ylabels, fontsize=12)

    # Add title
    plt.title("Correlation Matrix", fontsize=14)
```

```python
import scipy.stats as stats
def plot_single_correlation(treatment_data, outcome_data, treatment_name, outcome_name, correlation=None):
    fig, ax1 = plt.subplots()

    color = 'tab:red'
    ax1.set_xlabel('samples')
    ax1.set_ylabel(treatment_name, color=color)
    ax1.plot(treatment_data, color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    # Rotate x-axis labels vertically
    ax1.tick_params(axis='x', labelrotation=90)

    ax2 = ax1.twinx()  # instantiate a second Axes that shares the same x-axis

    color = 'tab:blue'
    ax2.set_ylabel(outcome_name, color=color)  # we already handled the x-label with ax1
    ax2.plot(outcome_data, color=color)
    ax2.tick_params(axis='y', labelcolor=color)

    rho, p_value = stats.spearmanr(treatment_data, outcome_data)
    correlation = rho
    if correlation is not None:
        plt.title(f"Correlation: {round(float(rho), 4)}")
    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.savefig(f"./figures/{treatment_name}-{outcome_name}.png")
```

```python
plot_single_correlation(cyanobacteria, xanthine, 'cyanobacteria', 'xanthine')
```

### Normalizing
For soil data: see what normalization does for experiment by normalizing each sample by maximum value in that sample (both microbes and metabolites)

Let's start with the microbes

```python
microbes_norm = microbes / microbes.max()
```

```python
microbes_norm.head()
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
plt.plot(microbes_norm.sum(axis=1))
ax.tick_params(axis='x', labelrotation=90)
```

Now for the metabolites

```python
metabolites_norm = metabolites / metabolites.max()
metabolites_norm.head()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
plt.plot(metabolites_norm.sum(axis=1))
ax.tick_params(axis='x', labelrotation=90)
```
