```python
# apply normalization to relative abundances
microbes = pd.read_csv('./data/abs_microbes.from_biom.tsv', sep='\t', index_col=0)
metabolites = pd.read_csv('./data/abs_metabolites.from_biom.tsv', sep='\t', index_col=0)
```

```python
# apply log transformation
# first increase all values with 1
microbes += 1
metabolites += 1

microbes = np.log(microbes)
metabolites = np.log(metabolites)

# normalize dataframes
microbes_normalized = normalize_df(microbes)
metabolites_normalized = normalize_df(metabolites)
metabolites_normalized = metabolites_normalized.T.reindex(microbes_normalized.T.index).T

# save transformed data
microbes_normalized.to_csv("./data/abs_microbes.normalized.tsv", sep="\t", index=True)
metabolites_normalized.to_csv("./data/abs_metabolites.normalized.tsv", sep="\t", index=True)
```

```python
import json

def to_json(input):
    return input.replace("'", '"').replace('array(', '').replace(')', '').replace('.]', '.0]').replace('[', '').replace(']', '')

rv = pd.DataFrame(columns = sensitivity.columns, index = sensitivity.index)

for i in sensitivity.columns:
    for j in sensitivity.index:
        sens_out = sensitivity[i][j]
        sens_dict = json.loads(to_json(sens_out))
        rv.at[j, i] = sens_dict['rv']

rv.dtypes
```
